```python
def salary(base, allowance=0, tax=0):
    if base < 0:
        return "Invalid Base Salary"

    total = base + allowance - tax
    return total

print(salary(20000, 5000, 2000))
```

```python
def cart_total(*prices, discount=0):
    total = 0

    for price in prices:
        total += price

    total = total - (total * discount / 100)

    return total

print(cart_total(100, 200, 300, discount=10))
```

```python
def reserve(order, inventory):
    for item in order:
        if item["name"] not in inventory:
            return "Item Not Found"

        if inventory[item["name"]] < item["qty"]:
            return "Insufficient Stock"

        inventory[item["name"]] -= item["qty"]

    return "Success"
```

```python
def stock_filter(inventory):
    in_stock = []
    out_stock = []

    for item in inventory:
        if inventory[item] > 0:
            in_stock.append(item)
        else:
            out_stock.append(item)

    return in_stock, out_stock
```

```python
def read_file(file):
    with open(file) as f:
        for line in f:
            yield line
```

```python
import json

def process_orders(orders):
    result = []

    for order in orders:
        if order["amount"] > 150:
            order["status"] = "processed"
            result.append(order)

    return json.dumps(result)
```

```python
import json

def read_json(file):
    with open(file) as f:
        data = f.read()

    return json.loads(data)
```
